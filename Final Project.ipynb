```python
%matplotlib inline
import pandas
```

```python
df = pandas.read_excel('s3://carroll-ia241/COVID19.xlsx')

df[:10]
```

## Data from Benin

```python
benin_data = df.loc[df['countriesAndTerritories'] == 'Benin']
benin_data
```

## Total Deaths

```python
total_deaths = df.groupby('countriesAndTerritories').sum()['deaths']

total_deaths.nlargest(5).plot.bar()
```

## Top 10 cases

```python
most_cases = df.groupby('countriesAndTerritories').sum()['cases']
most_cases.nlargest(10).hist()
```

## USA deaths by month

```python
usa_data = df.loc[df['countriesAndTerritories'] == 'United_States_of_America']

usa_data.plot.scatter(x = 'month', y = 'deaths')
```

## Cases by Continent

```python
eu_cases = df.groupby('continentExp').sum()['cases']

eu_cases.plot.bar()
```
